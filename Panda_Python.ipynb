```python
##Função para colorir a tabela de acordo com o que você quer:

def colorir_tab_red(value):
    color = 'red' if value < 600000 else 'black'
    return 'color: %s' %color

df[['preco']].head(10).style.applymap(colorir_tab_red)
```
